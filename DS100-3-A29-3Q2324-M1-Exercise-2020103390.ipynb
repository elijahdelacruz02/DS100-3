```python
new_data = new_data.reset_index(drop=True)
new_data.to_csv('abs-change-energy-consumption.csv')
```

```python
pd.set_option('display.max_rows', None)
new_data = new_data.loc[new_data['Year'] >= 1991]
new_data = new_data.reset_index(drop=True)
new_data
```

```python
# Calculate the relative change
new_data['Relative_change'] = new_data['Annual change in primary energy consumption (TWh)'].pct_change() * 100

# Calculate the absolute change
new_data['Absolute_change'] = new_data['Annual change in primary energy consumption (TWh)'].diff().abs()
```

```python
print("Importing data")
import re

df = new_data.loc[new_data['Entity'].str.contains('Philippines', regex=True)]
df = df.rename(columns={'Entity': 'Entity','Year':'Year', 'Annual change in primary energy consumption (TWh)': 'Annual_change(Twh)',
                        'Absolute_change': 'Absolute_change(Twh)', 'Relative_change': 'Relative_change (%)'})
df = df.drop(columns=['Code'])
```

```python
mod_data = pd.merge(new_data,df)
mod_data[['Entity','Year','Annual_change(Twh)','Absolute_change(Twh)', 'Relative_change (%)']]
```
